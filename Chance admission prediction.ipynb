<h1>Chance of Admission for Higher Studies</h1>
Predict the chances of admission of a student to a Graduate program based on:

1. GRE Scores (290 to 340)
2. TOEFL Scores (92 to 120)
3. University Rating (1 to 5)
4. Statement of Purpose (1 to 5)
5. Letter of Recommendation Strength (1 to 5)
6. Undergraduate CGPA (6.8 to 9.92)
7. Research Experience (0 or 1)
8. Chance of Admit (0.34 to 0.97)

```python
#import library
import pandas as pd
```

```python
#import data
admission = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Admission%20Chance.csv')
```

```python
# Target(y) and Feature(X)
y = admission['Chance of Admit ']
X = admission.drop(['Chance of Admit '],axis = 1)
```

```python
y.shape,X.shape
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y ,random_state = 2529)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
# Select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# train model or fit model
y_pred= model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# model predict
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#model accurecy
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error
```

```python
mean_absolute_error(y_pred,y_test)
```

```python
mean_absolute_percentage_error(y_pred,y_test)
```

```python
mean_squared_error(y_pred,y_test)
```
