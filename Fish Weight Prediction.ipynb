<p>With a dataset of fish species, with some of it characteristic like it vertical, diagonal, length, height, and width. We will try to predict the weight of the fish based on their characteristic. We will use Linear Regression Method to see whether the weight of the fish related to their characteristic.</p

<ol>
    <li>Species: Species name of fish</li>
    <li>Weight: Weight of fish in gram</li>
    <li>Length1: Vertical length in cm</li>
    <li>Length2: Diagonal length in cm</li>
    <li>Length3: Cross length in cm</li>
    <li>Height: Height in cm</li>
    <li>Width: Diagonal width in cm</li>
</ol>

```python
# import library
import pandas as pd
```

```python
#import data
fish = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Fish.csv')
```

```python
fish.head()
```

```python
fish.info()
```

```python
fish.describe()
```

```python
# target(y) and feature(X)
y = fish['Weight']
X = fish[['Height', 'Width', 'Length1',
       'Length2', 'Length3']]
```

```python
y.shape,X.shape
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test =train_test_split(X,y, train_size=0.7, random_state = 2529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
#step 05: Select model (best fit)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#step 06: train and fit model
model.fit(X,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# step 07 : Model predict
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#step 08 : Accurecy
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error
```

```python
mean_squared_error(y_pred, y_test)
```

```python
mean_absolute_error(y_pred, y_test)
```

```python
mean_absolute_percentage_error(y_pred, y_test)
```
