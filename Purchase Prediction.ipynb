<h1>Customer Purchase Prediction & Effect of Micro-Numerosity</h1>

```python
#import library
import pandas as pd
```

```python
#import datasete
purchase = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Customer%20Purchase.csv')
```

```python
purchase.head()
```

```python
purchase.info()
```

```python
purchase.describe()
```

```python
purchase.columns
```

```python
# Define target(y) and Feature(X)
y = purchase['Purchased']
X = purchase[['Customer ID', 'Age', 'Gender', 'Education', 'Review']]
```

```python
X.shape,y.shape
```

```python
#Encoding categorical variables
X.replace({'Review':{'Average':0,'Poor':1,'Good':2 }},inplace = True)
X.replace({'Education':{'School':0,'UG':1,'PG':2}},inplace = True)
X.replace({'Gender':{'Female':0,'Male':1}},inplace = True)
```

```python
X.head()
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.7,random_state = 2529)
# X_train
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# select model(least model select)
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
# model train of model fit
model.fit(X_train,y_train)
```

```python
# model predict
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#model accurecy
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```
